```python
import os
import pandas as pd
from pyspark.sql import SparkSession

# Inicia la sesión de Spark
spark = SparkSession.builder \
    .appName("DeltaTableCreation") \
    .config("spark.jars.packages", "io.delta:delta-core_2.12:2.4.0") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

# Ruta a los CSV generados y a la tabla Delta
csv_folder = "./Data/csv"
delta_table_path = "./Data/delta_table"

# Leer todos los CSV en un DataFrame de Pandas
csv_files = [os.path.join(csv_folder, file) for file in os.listdir(csv_folder) if file.endswith(".csv")]

# Combina todos los CSV en un único DataFrame de Pandas
pandas_df = pd.concat([pd.read_csv(csv_file) for csv_file in csv_files], ignore_index=True)

# Convierte el DataFrame de Pandas a Spark
spark_df = spark.createDataFrame(pandas_df)

# Escribe el DataFrame en formato Delta
spark_df.write.format("delta").mode("overwrite").save(delta_table_path)

print(f"Tabla Delta creada en: {delta_table_path}")
```

```python
print(spark.version)
```

```python
# Leer la tabla Delta
delta_df = spark.read.format("delta").load(delta_table_path)

# Muestra las primeras filas de la tabla
delta_df.show()

# Muestra el esquema de la tabla Delta
delta_df.printSchema()
```
